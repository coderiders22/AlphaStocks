```python
import pandas as pd
import matplotlib.pyplot as plt
from mplfinance.original_flavor import candlestick_ohlc
import matplotlib.dates as mdates

# Filter DataFrame for the "WIPRO" stock
wipro_data = rm[rm['SYMBOL'] == 'WIPRO']

# Convert the 'TIMESTAMP' column to datetime format
wipro_data['TIMESTAMP'] = pd.to_datetime(wipro_data['TIMESTAMP'])

# Set 'TIMESTAMP' as the index
wipro_data.set_index('TIMESTAMP', inplace=True)

# Resample the data to a daily frequency and select OHLC data
ohlc_data = wipro_data.resample('D').agg({'OPEN': 'first', 'HIGH': 'max', 'LOW': 'min', 'CLOSE': 'last'})

# Reset index to make 'TIMESTAMP' a column
ohlc_data.reset_index(inplace=True)

# Convert 'TIMESTAMP' to matplotlib date format
ohlc_data['TIMESTAMP'] = ohlc_data['TIMESTAMP'].map(mdates.date2num)

# Plot candlestick chart
plt.figure(figsize=(10, 6))
ax = plt.subplot()
candlestick_ohlc(ax, ohlc_data.values, width=0.6, colorup='g', colordown='r')

# Format x-axis as dates
ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.title('Candlestick Chart for WIPRO Stock')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Get unique stock symbols from the dataframe
symbols = rm['SYMBOL'].unique()

# Prepare an empty dictionary to store LSTM models and predictions for each symbol
models = {}
predictions = {}

# Iterate over each symbol
for symbol in symbols:
    # Filter data for the current symbol
    symbol_data = rm[rm['SYMBOL'] == symbol]
    
    # Convert DataFrame to NumPy array
    symbol_array = symbol_data[['OPEN', 'HIGH', 'LOW', 'CLOSE', 'LAST', 'PREVCLOSE', 'TOTTRDQTY', 'TOTTRDVAL', 'TOTALTRADES']].values
    
    # Reshape the input data
    symbol_array = np.reshape(symbol_array, (symbol_array.shape[0], symbol_array.shape[1], 1))
    
    # Define LSTM model architecture
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=(symbol_array.shape[1], 1)))
    model.add(LSTM(50))
    model.add(Dense(1))  # One output for predicted closing price
    
    # Train the LSTM model on the data for the current symbol
    model.compile(loss='mse', optimizer='adam')
    model.fit(symbol_array[:-1], symbol_array[1:, 3], epochs=50)  # Assuming the CLOSE column is the target to predict
    
    # Save the trained model for the current symbol
    models[symbol] = model
    
    # Use the trained LSTM model to make predictions for the current symbol
    symbol_prediction = model.predict(symbol_array[-1:])[:, 0]
    
    # Save the predictions for the current symbol
    predictions[symbol] = symbol_prediction

# Now you have models and predictions for each symbol
# You can access them using the respective symbol as the key in the dictionaries
```

```python
from sklearn.model_selection import train_test_split

# Get unique stock symbols from the dataframe
symbols = rm['SYMBOL'].unique()

# Prepare dictionaries to store LSTM and ARIMA models and predictions for each symbol
lstm_models = {}

```

```python
from sklearn.preprocessing import MinMaxScaler

for symbol in symbols:
    # Filter data for the current symbol
    symbol_data = rm[rm['SYMBOL'] == symbol]
    
    # Extract features and target
    features = symbol_data[['OPEN', 'HIGH', 'LOW', 'CLOSE', 'LAST', 'PREVCLOSE', 'TOTTRDQTY', 'TOTTRDVAL', 'TOTALTRADES']].values
    target = symbol_data['CLOSE'].values
    
    # Scale the features to the range [0, 1]
    scaler = MinMaxScaler(feature_range=(0, 1))
    features_scaled = scaler.fit_transform(features)
    
    # Reshape the input data
    features_scaled = np.reshape(features_scaled, (features_scaled.shape[0], features_scaled.shape[1], 1))
    
    
    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(features_scaled[:-1], target[1:], test_size=0.2, shuffle=False)  # Assuming 80% training and 20% testing
    
    # Define LSTM model architecture
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=(features_scaled.shape[1], 1)))
    model.add(LSTM(50))
    model.add(Dense(1))  # One output for predicted closing price
    
    # Train the LSTM model on the training data for the current symbol
    model.compile(loss='mse', optimizer='adam')
    model.fit(X_train, y_train, epochs=100)
    
    # Save the trained LSTM model for the current symbol
    lstm_models[symbol] = {'model': model, 'scaler': scaler}
```

```python


# Prepare dictionaries to store buy/sell/hold signals and prices for each stock in the portfolio
portfolio_signals = {}
portfolio_prices = {}
portfolio_volatility = {}
```

```python
for symbol in portfolio_symbols:
    lstm_prediction = lstm_predictions[symbol]
    # Prompt the user to input the risk tolerance level for the current stock
    risk_tolerance = input(f"Enter risk tolerance level for {symbol} (high/medium/low): ").lower()
    # Get the buy and sell thresholds for the current risk tolerance level
    buy_threshold = risk_tolerance_levels[risk_tolerance]['buy_threshold']
    sell_threshold = risk_tolerance_levels[risk_tolerance]['sell_threshold']
    # Determine buy/sell/hold signal based on LSTM prediction and thresholds
    if lstm_prediction >= buy_threshold:
        signal = 'BUY'
    elif lstm_prediction <= sell_threshold:
        signal = 'SELL'
    else:
        signal = 'HOLD'
    # Save the signal for the current stock
    portfolio_signals[symbol] = signal
    # Save the current price for the current stock
    portfolio_prices[symbol] = current_prices[symbol]
    # Save the volatility for the current stock
    portfolio_volatility[symbol] = volatility[symbol]

# Calculate the total predicted risk and total volatility
total_prediction = sum(lstm_predictions[symbol] for symbol in portfolio_symbols)
total_volatility = sum(portfolio_volatility.values())

# Adjust capital allocation based on both predicted risk and volatility
allocation = {symbol: (capital * (lstm_predictions[symbol] / total_prediction) * (portfolio_volatility[symbol] / total_volatility)) for symbol in portfolio_symbols}
```

```python
for symbol in portfolio_symbols:
    if portfolio_signals[symbol] == 'BUY':
        # Calculate the number of shares to buy based on allocated capital and current price
        num_shares_to_buy = allocation[symbol] // portfolio_prices[symbol]
        print(f"For symbol {symbol}: {portfolio_signals[symbol]} {num_shares_to_buy} shares.")
    elif portfolio_signals[symbol] == 'SELL':
        # Calculate the number of shares to sell based on allocated capital and current price
        num_shares_to_sell = allocation[symbol] // portfolio_prices[symbol]
        print(f"For symbol {symbol}: {portfolio_signals[symbol]} {num_shares_to_sell} shares.")
    else:
        print(f"For symbol {symbol}: {portfolio_signals[symbol]}")
```

```python
import streamlit as st
import pickle

# Load your saved portfolio optimization model from the pickle file
@st.cache(allow_output_mutation=True)
def load_model():
    with open('/kaggle/working/current_prices.pkl', 'rb') as f:
        model = pickle.load(f)
    return model

# Main Streamlit app
def main():
    st.title('Portfolio Optimization App')
    
    # Load the model
    model = load_model()
    
    # User inputs
    capital = st.number_input('Enter your capital:', min_value=0.0)
    risk_tolerance = st.slider('Select your risk tolerance (1 - Low, 2 - Medium, 3 - High):', 1, 3)
    
    # Make predictions with the model
    if st.button('Optimize Portfolio'):
        # Assuming you have a function to make predictions with your model
        portfolio_allocations = model.predict(capital, risk_tolerance)
        
        # Display the recommended portfolio allocations
        st.write('Recommended portfolio allocations:')
        for symbol, allocation in portfolio_allocations.items():
            st.write(f'Symbol: {symbol}, Allocation: {allocation}')

if __name__ == '__main__':
    main()
```
